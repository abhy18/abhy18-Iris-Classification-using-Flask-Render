```python
import numpy as np
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import LabelEncoder
```

```python
df = pd.read_csv('/content/IRIS.csv')
df
```

```python
lbl = LabelEncoder()
df['species'] = lbl.fit_transform(df['species'])
```

```python
x = df.drop('species',axis=1)
y = df['species']
```

```python
df
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test ,y_train,y_test = train_test_split(x,y,test_size=.25)
```

```python
iris_clf = RandomForestClassifier(n_estimators=150,criterion = 'gini',max_depth=14,max_samples=50)
iris_clf.fit(x_train,y_train)
```

```python
y_pred = iris_clf.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_pred,y_test)
```

```python
import pickle as pkl
pkl.dump(iris_clf,open('iris_classifier.pkl','wb'))
```
